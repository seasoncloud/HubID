```python
# %%
import numpy as np
import scanpy as sc
from matplotlib import pyplot as plt
import pandas as pd
import anndata as ad
import tensorflow as tf
import pickle
from tensorflow_probability import math as tm
tfk = tm.psd_kernels
```

```python
from nsfpaper.models import sf
from nsfpaper.utils import misc,preprocess,training,postprocess,visualize
```

# Baysor data

```python
counts = pd.read_csv("../data/sub7_cell_by_gene.csv")
X = pd.read_csv("../data/sub7_location.csv")
X = X.iloc[:,[1,2]].to_numpy()
```

```python
adata = ad.AnnData(counts)
adata.obsm["spatial"] = X
adata.layers["counts"] = counts.to_numpy()
```

```python
D,Dval = preprocess.anndata_to_train_val(adata, layer="counts", train_frac=0.9,
                                         flip_yaxis=False)
Ntr,J = D["Y"].shape
Xtr = D["X"]

Dtf = preprocess.prepare_datasets_tf(D, Dval = Dval)
```

```python
L = 5 #number of components
Z = misc.kmeans_inducing_pts(Xtr, 2000)
M = Z.shape[0] #number of inducing points
ker = tfk.MaternThreeHalves
```

```python
fit = sf.SpatialFactorization(J, L, Z, psd_kernel=ker, nonneg=True, lik="poi")
fit.init_loadings(D["Y"], X=Xtr, sz=D["sz"], shrinkage=0.3)
tro = training.ModelTrainer(fit)
%time tro.train_model(*Dtf, status_freq=50) #about 3 mins
visualize.plot_loss(tro.loss)
```

```python
listout = [fit,tro.loss,tro.wtime]

def save_object(obj, filename):
    with open(filename, 'wb') as outp:  # Overwrites any existing file.
        pickle.dump(obj, outp, pickle.HIGHEST_PROTOCOL)

save_object(listout,"nsf_result/sub7_2000ip.pkl")
```

```python
#%% NSF postprocessing and visualization
insf = postprocess.interpret_nsf(listout[0],Xtr,S=100,lda_mode=False)
Fplot = insf["factors"]
fig,axes=visualize.multiheatmap(Xtr, Fplot, (1,5), cmap="Blues", s = 0.1, figsize = [18,5])
```
