```r

spatial = read.csv("data/spatial_sample1.csv", row.names = 1)

labels = read.csv("data/kmean_anno_sample1.csv", row.names = 1)
```

```r
head(labels)
```

```r
library(ggplot2)
options(repr.plot.width=20, repr.plot.height=10)
ggplot(data.frame(x = spatial[,1], y = spatial[,2], label = labels[,1]), aes(x = x, y = y, col = label))+
geom_point(cex = 0.05)+
scale_color_manual(values = c('#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#e6beff', '#9a6324'))+
guides(color = guide_legend(override.aes = list(size = 2)))+
theme(axis.text = element_blank(),
axis.ticks = element_blank())
```
